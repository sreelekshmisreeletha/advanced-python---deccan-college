```python
data = 'Java is <replaced> by Python Programming.'
```

```python
data = re.sub("<.*>", 'in this century', data)
```

```python
data
```

```python
data = 'Java is <replaced> by Python Programming <language>.'
```

```python
data = re.sub("<.*>", 'in this century', data)
```

```python
data
```

```python
data = 'Java is <replaced> by Python Programming <language>.'
data = re.sub("<.*?>", 'in this century', data)
data
```

```python
data = 'Java is <replaced> by Python Programming <language>.'
data = re.sub("<.*?>", ' ', data)
data
```

```python
txt = 'There were only 5 students present out of 26'
```

```python
re.sub('\d', '', txt)
```

```python
re.sub('\d', '', txt).split()
```

```python
re.sub('[^\d]', '', txt)
```

```python
re.sub('[^\d]', ' ', txt).split()
```

```python
re.sub('[a-zA-z]', ' ', txt)
```

```python
re.sub('[^a-zA-z]', ' ', txt)
```

```python
re.sub('[^a-zA-z]', ' ', txt).split()
```

```python
re.sub('\W+', ' ', txt)
```

```python
pattern = "<title.*?>.*?</title.*?>"
match_results = re.search(pattern, html_text, re.IGNORECASE)
title = match_results.group()
title = re.sub("<.*?>", "", title)# Remove HTML tags
print(title)
```
